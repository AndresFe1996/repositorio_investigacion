```python
### limpieza de datos 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# abrir archivo

data=pd.read_csv("C:/Users/Home/Repositorio_investigacion/Data/Gran Encuesta Alianza ESC Octubre 2023 (Respuestas) - Respuestas de formulario 1.csv")
data
```

```python
# convertir columna en fecha
data["Marca temporal"]=pd.to_datetime(data["Marca temporal"])

# limpieza de datos
```

```python
# analizando variables individualmente

# mapeamos cada valor de texto a un valor numerico

mapping={cat: idx for idx, cat in enumerate(data['3. Nacionalidad'].unique())}
data['cat_naci']=data['3. Nacionalidad'].map(mapping)
```

```python
## graficamos

plt.figure(figsize=(10,10))

sns.boxplot(x='cat_naci',y='3. Nacionalidad',data=data)
plt.show()
```

```python
## arreglando la tipografica de Nacionalidad
## usamos el metodo replace y su variable regex para que convierta todas las casualidad en Masculino y Femenino
data['3. Nacionalidad']=data['3. Nacionalidad'].replace({"Mexicana":"México","Chileno":"chile",
                                                                  "Española":"España","mexicana":"México",
                                                        "Chilena":"chile","Mexico":"México",
                                                        "España ":"España"},regex=True)

```

```python
## arreglado
plt.figure(figsize=(10,10))

sns.boxplot(x='cat_naci',y='3. Nacionalidad',data=data)
plt.show()
```

```python
# pregunta 2
# 2. ¿Cuál es tu puntuación actual en el juego?

puntajes=data["2. ¿Cuál es tu puntuación actual en el juego?"]

sns.histplot(puntajes, kde=True)  # Utilizar seaborn para un histograma más estilizado y la estimación de la densidad del kernel (kde)
plt.title('Histograma y KDE')
plt.show()
```

```python
# grado de dispersion de puntos tecnologicos con boxplot
# pregunta 7

q1 = data['7. ¿Cuál es tu nivel de producción de puntos de investigación por hora? (el dato lo sacas dando clic en el consejero académico por si no lo sabes, si tiene decimales coloca (.) como separador ejemplo: 701.78)'].quantile(0.25)
q2 = data['7. ¿Cuál es tu nivel de producción de puntos de investigación por hora? (el dato lo sacas dando clic en el consejero académico por si no lo sabes, si tiene decimales coloca (.) como separador ejemplo: 701.78)'].quantile(0.50)  # mediana
q3 = data['7. ¿Cuál es tu nivel de producción de puntos de investigación por hora? (el dato lo sacas dando clic en el consejero académico por si no lo sabes, si tiene decimales coloca (.) como separador ejemplo: 701.78)'].quantile(0.75)

# Crear el box plot con Seaborn
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
ax = sns.boxplot(x=data['7. ¿Cuál es tu nivel de producción de puntos de investigación por hora? (el dato lo sacas dando clic en el consejero académico por si no lo sabes, si tiene decimales coloca (.) como separador ejemplo: 701.78)'])

# Agregar líneas para cuartiles y mediana
ax.axhline(q1, linestyle='--', color='red', label='Q1')
ax.axhline(q2, linestyle='--', color='green', label='Q2 (Mediana)')
ax.axhline(q3, linestyle='--', color='blue', label='Q3')

# Agregar etiquetas
ax.set_title('Box Plot con Cuartiles')
ax.set_xlabel('Datos')

# Mostrar el gráfico
plt.legend()
plt.show()
```
